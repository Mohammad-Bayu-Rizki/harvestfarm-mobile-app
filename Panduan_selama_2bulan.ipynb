##### **Minggu 2: Manipulasi Data dengan Pandas & NumPy**

  * **Tujuan:** Mampu memuat, membersihkan, dan mengubah data tabular menggunakan *library* paling fundamental.
  * **Konsep Kunci:**
      * **Pandas:** DataFrame, Series, membaca file (CSV, Excel), seleksi data (indexing, slicing), menangani data yang hilang (*missing values*).
      * **NumPy:** Array, operasi matematis pada array (vektorisasi).
  * **Analogi:** Bayangkan Anda adalah seorang koki. **Pandas** adalah meja dapur super canggih Anda (DataFrame) tempat Anda menata, memotong, dan membersihkan semua bahan (data). **NumPy** adalah set pisau dan alat masak presisi tinggi yang melakukan semua perhitungan dan pemotongan dengan sangat cepat dan efisien.
  * **Contoh Kode Singkat (Pandas):**

```
import pandas as pd
    # Membuat DataFrame dari dictionary
    data = {'Mahasiswa': ['Ani', 'Budi', 'Joko'], 'Nilai': [85, 92, 78]}
    df = pd.DataFrame(data)

    # Menampilkan DataFrame
    print(df)

    # Menghitung rata-rata nilai
    rata_rata_nilai = df['Nilai'].mean()
    print(f"Rata-rata nilai mahasiswa: {rata_rata_nilai}")
```

* **Sumber Belajar:**
      * **Dokumentasi Praktis:** [10 Minutes to pandas](https://pandas.pydata.org/pandas-docs/stable/user_guide/10min.html). Ini adalah ringkasan yang bagus dari dokumentasi resmi.
      * **Buku Anda:** Anda sudah punya `numpy ML.pdf`. Mulailah membaca dari bagian "Pandas Introduction" (halaman 7) dan ikuti contoh-contohnya.
  * **Proyek Mini:**

        1.  Berapa jumlah total penumpang?
        2.  Berapa rata-rata usia penumpang?
        3.  Berapa banyak penumpang di setiap kelas tiket (`Pclass`)?

#### **Minggu 3-4: Fase Eksplorasi (Exploration) - Menemukan Cerita dalam Data**

##### **Minggu 3: Visualisasi Data & Statistik Dasar**

  * **Tujuan:** Mampu membuat grafik yang informatif dan memahami statistik deskriptif untuk merangkum data.

  * **Sumber Belajar:**

##### **Minggu 4: Pengenalan Machine Learning (ML)**

  * **Tujuan:** Memahami konsep dasar ML dan membangun model prediksi pertama menggunakan Scikit-Learn.
  * **Konsep Kunci:** Supervised vs Unsupervised Learning, Regresi vs Klasifikasi, *Train-Test Split*, Model Evaluation (Akurasi).
  * **Analogi:** **Machine Learning** itu seperti mengajari komputer dengan contoh. Untuk membedakan email **spam** dan **bukan spam** (Klasifikasi), kita berikan ribuan contoh email yang sudah dilabeli. Komputer akan belajar polanya sendiri. Saat ada email baru, ia bisa menebak labelnya dengan probabilitas tinggi.
  * **Sumber Belajar:**
      * **Buku Anda:** Mulai baca **Bab 1** dari buku `Hands-On Machine Learning with Scikit-Learn...`. Bab ini akan memandu Anda melalui proyek *end-to-end* pertama Anda.
  * **Proyek Portofolio \#1 (Klasifikasi Sederhana):**
      * **Dataset:** [Iris Species Dataset](https://www.kaggle.com/datasets/uciml/iris).
      * **Tujuan:** Membuat model yang bisa memprediksi spesies bunga Iris berdasarkan ukuran kelopak dan mahkotanya.
      * **Langkah:**
        1.  Muat data dengan Pandas.
        2.  Lakukan *Exploratory Data Analysis* (EDA) dengan visualisasi untuk melihat hubungan antar fitur.
        3.  Pisahkan data menjadi set latihan dan set pengujian (`train_test_split` dari Scikit-Learn).
        4.  Latih model sederhana seperti `LogisticRegression` atau `KNeighborsClassifier`.
        5.  Evaluasi akurasi model Anda.
        6.  **PENTING:** Buat *repository* baru di [GitHub](https://github.com/) dan letakkan *notebook* proyek ini di sana. Inilah awal dari portofolio Anda\!

### **Bulan 2: Pendalaman & Proyek Serius**

#### **Minggu 5-8: Fase Penguasaan (Mastery) - Membangun Keahlian Siap Industri**

##### **Minggu 5-6: Algoritma ML & Feature Engineering**

  * **Tujuan:** Mempelajari algoritma yang lebih kuat dan seni menciptakan fitur yang baik untuk model.
  * **Konsep Kunci:** *Decision Trees*, *Random Forest*, *Gradient Boosting*, *Feature Engineering*, *Cross-Validation*.
  * **Analogi:** Jika *Logistic Regression* adalah mobil keluarga, maka **Random Forest** dan **Gradient Boosting** (seperti XGBoost) adalah mobil balap F1. Mereka jauh lebih bertenaga dan akurat. **Feature Engineering** adalah seni memodifikasi mesin dan aerodinamika (data input) agar mobil balap Anda melaju lebih kencang.
  * **Sumber Belajar:**
      * **Lanjutkan Buku:** `Hands-On Machine Learning...` (Bab 2, 3, 6, dan 7). Buku ini membahas semua topik tersebut dengan sangat baik.
      * **Video:** Cari "StatQuest Random Forest" dan "StatQuest XGBoost" di YouTube.

##### **Minggu 7: Lompatan ke Deep Learning**

  * **Tujuan:** Memahami kapan dan mengapa menggunakan Deep Learning, serta membangun Jaringan Saraf Tiruan (Neural Network) pertama.
  * **Konsep Kunci:** *Neural Networks*, *Layers* (Dense), *Activation Functions* (ReLU, Sigmoid), Keras API.
  * **Analogi:** Jika ML tradisional hebat dalam data terstruktur (tabel Excel), **Deep Learning** bersinar pada data tidak terstruktur seperti **gambar, suara, dan teks**. Ini seperti memiliki otak buatan yang dapat mengenali pola-pola yang sangat kompleks secara otomatis.
  * **Sumber Belajar:**
      * **Buku Anda (Sangat Direkomendasikan):** `Deep Learning with Python` oleh François Chollet (pencipta Keras). Mulailah membaca Bab 1-3. Ini adalah pengantar terbaik untuk Deep Learning.
      * **Video:** [3Blue1Brown - Neural Networks](https://www.youtube.com/playlist?list=PLZHQObOWTQDNU6R1_67000Dx_ZCJB-3pi) untuk intuisi visualnya.
